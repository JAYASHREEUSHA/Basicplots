Importing libraries

```
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```


Importing Dataset


```
a = pd.read_csv('Book1.csv') 
print(a) 
```

Plotting Line graph

```
a.plot() 
plt.title("line graph") 
plt.show()
```

Plotting Scatter graph

```
a.plot(kind="scatter", x = "Pulse", y = "Calories") 
plt.xlabel = ('Pulse') 
plt.ylabel = ('calories') 
plt.title('Scatter plot') 
plt.show()
```

Plotting Histogram

```
a["Duration"].plot(kind = 'hist') 
plt.title('Histogram') 
plt.show() 
```

Plotting Bar graph

```
df = pd.read_csv('book2.csv') 
x = df.iloc[:,:-1].values 
y = df.iloc[:,:-1].values 
x_axis = np.array([x]) 
y_axis = np.array([y]) 
re_shape = np.reshape(x_axis,16) 
print(re_shape) 
re_shape1 = np.reshape(y_axis,16) 
print(re_shape1) 
plt.bar(re_shape, re_shape1) 

plt.show() 
```

Plotting Pie Chart

```
pd.read_csv('Book1.csv') 
df['Calories'].plot(kind='pie') 
plt.title('Pie Chart') 
plt.show()
```
